## Catagory Web Request 

Get the url of each fountain pen page from the fountain pen search catagory. 

```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bsoup
import re
```

```python
pen_cards = []
i = 1
while i < 58:
    page_url = "https://www.gouletpens.com/collections/all-fountain-pens/products?page="+str(i)
    search_content = bsoup(requests.get(page_url).text, 'lxml')
    pen_cards += search_content.find_all("div",{"class":"product-fallback__item grid__item one-half large-up--one-third"})
    i = i + 1
```

```python
pen_url = list(map(lambda x: "https://www.gouletpens.com" + x.find('a').get('href'), pen_cards))
```

```python
pen_df = pd.DataFrame(pen_url,columns=["url"],index=range(676))
```

```python
pen_df.to_csv("fountain_pen_dataset.csv")
```

### add basic introduction

```python
PC, N, IS, P, NS = [],[],[],[],[]
for url in pen_df.url:
    soup = bsoup(requests.get(url).text, 'lxml').find("div",{"data-section-type":"product"})
    PC.append(soup.find("span",{"class":"variant-sku"}).text)
    N.append(re.sub("on sale","",soup.find("h1",{"class":"product-info__title"}).text).strip())
    if soup.find("p",{"class":"in-stock"}):
        IS.append(True)
    else:
        IS.append(False)
    P.append(soup.find("div",{"class":"product-price"}).text.strip())
    NS.append(list(map(lambda x: x.get("value"),soup.find("select",{"aria-label":"Product Variant Selector"}).find_all("option"))))
    print(len(N))
```

```python
pens_df["Product_Code"] = PC
pens_df["Name"] = N
pens_df["In_Stock"] = IS
pens_df["Price"] = P
pens_df["Nib_Size"] = NS
```

### add specs list discription

```python
## Specs_List
for i in range(136,pens_df.shape[0]):
    soup = bsoup(requests.get(pens_df.url[i]).text, 'lxml').find("div",{"data-section-type":"product"})
    try:
        col_names = map(lambda x: re.sub(" - | ","_",re.sub(r"\n.*","",x.text.strip())),soup.find("dl",{"class":"specs-list"}).find_all("dt"))
        col_spec = map(lambda x: x.text.strip(),soup.find("dl",{"class":"specs-list"}).find_all("dd"))
        for idx in range(len(col_names)):
            pens_df.set_value(i, col_names[idx], col_spec[idx])
    except:
        print(i)
```

### add basic json information

```python
pens_df = pd.read_csv("fountain_pen_dataset.csv").drop(["Unnamed: 0","Unnamed: 0.1"],axis=1)
pens_df.head()
```

```python
import json
jsons = []
for i in range(pens_df.shape[0]):
    soup = bsoup(requests.get(pens_df.url[i]).text, 'lxml').find("div",{"data-section-type":"product"})
    try:
        dic = json.loads(soup.find("script",{"type":"application/json"}).text)
        jsons.append(dic)
    except:
        jsons.append(list())
        print("-----"+str(i))
    print(i)
```

```python
# json_df = pd.DataFrame(jsons)
# json_df
null_json = jsons[1]
for i in null_json:
    null_json[i] = None
for i in range(len(jsons)):
    if not jsons[i]:
        jsons[i] = null_json
        
json_df = pd.DataFrame(jsons)
```

```python
pens_df = pd.concat([pens_df, json_df], axis=1, join_axes=[pens_df.index])
```

```python
pens_df.to_csv("fountain_pen_dataset.csv")
```
